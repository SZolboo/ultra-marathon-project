```python
# Output: 55345
# Keeping (50km, & 100km) & from 2012 to 2022 using the isin()
aus_athletes_subset = aus_athletes[(aus_athletes['Event distance/length'].isin(['50km', '100km'])) & (aus_athletes['Year of event'].between(2012, 2022))]
aus_athletes_subset
```

The country abv was at the end of the "Event name" column. I want to subset to keep only events from Australia.

```python
# Now, the output is: 52256
aus_races = aus_athletes_subset[aus_athletes_subset['Event name'].str.endswith('(AUS)')]
aus_races
```

## Data Cleaning

```python
# Check duplicate values
aus_races[aus_races.duplicated() == True]
```

```python
# Checking NULL values
aus_races.isna().sum()
```

```python
# Fill NaN values with a placeholder of 0
aus_races['Athlete year of birth'] = aus_races['Athlete year of birth'].fillna(0)

# Convert the column to integers
aus_races['Athlete year of birth'] = aus_races['Athlete year of birth'].astype(int)
```

```python
# Dropping the age that is 0
aus_races = aus_races[aus_races['Athlete year of birth'] != 0]
```

```python
# Output now is: 37574
aus_races
```

```python
# Cleaning the "Athlete performance" column.
aus_races['Athlete performance'] = aus_races['Athlete performance'].str.replace(' h', '')
aus_races
```

```python
# Cleaning the "Event name" column.
# All of the events are now from Australia.
aus_races.loc[:, 'Event name'] = aus_races['Event name'].str.replace(' (AUS)', '')
aus_races
```

```python
# 175 unique Events are in the dataset.
aus_races['Event name'].nunique()
```

```python
aus_races['Athlete age'] = aus_races['Year of event'] - aus_races['Athlete year of birth']
aus_races
```

```python
# Dropped unnecessary columns
aus_races.drop(['Athlete club', 'Athlete country', 'Athlete age category'], axis = 1, inplace = True)
```

```python
aus_races
```

```python
# Resetting the index
aus_races.reset_index(drop=True, inplace=True)
```

```python
# Convert the "Athlete average speed" column to numeric type
aus_races['Athlete average speed'] = pd.to_numeric(aus_races['Athlete average speed'], errors='coerce')

# Round the numeric values in the "Athlete average speed" column to two decimal places
aus_races['Athlete average speed'] = aus_races['Athlete average speed'].round(2)
```

```python
dates = aus_races['Event dates']

# Split the date strings by '.' separator
date_parts = dates.str.split('.')

# Extract the month part from each date string
months = date_parts.apply(lambda x: int(x[1]))

# Handle cases where multiple days are mentioned
# If there are multiple days, take the last day mentioned
for i, date_part in enumerate(date_parts):
    if len(date_part) > 2:
        months[i] = int(date_part[-2])
```

```python
# Creating a new month column after extracting the month value from the Event date column.
aus_races['Event months'] = months
```

### Summer vs. Winter comparisons

```python
plt.figure(figsize=(12, 6))

# Define a lambda function to map months to seasons
def map_to_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

# Apply the lambda function to create a new column 'Season'
aus_races['Season'] = aus_races['Event months'].apply(lambda x: map_to_season(x))

# Calculate average speeds for summer and winter seasons
summer_average_speed = aus_races.loc[aus_races['Season'] == 'Summer', 'Athlete average speed'].mean()
winter_average_speed = aus_races.loc[aus_races['Season'] == 'Winter', 'Athlete average speed'].mean()

# Create a bar plot to visualize the average speeds for summer and winter
sns.barplot(x = ['Summer', 'Winter'], y = [summer_average_speed, winter_average_speed], palette = "coolwarm")
plt.xlabel('Season')
plt.ylabel('Average Speed')
plt.title('Average Speed Comparison between Summer and Winter Seasons')
plt.show()
```

```python
# Difference in speed for the both 50km & 100km for male and female. 
aus_races.groupby(['Event distance/length', 'Athlete gender'])['Athlete average speed'].mean()
```

```python
# Group by 'Event distance/length' and 'Athlete gender' and calculate the mean average speed
avg_speed_by_distance_gender = \
    aus_races.groupby(['Event distance/length', 'Athlete gender'])['Athlete average speed'].mean().reset_index()

# Plot the grouped bar plot
plt.figure(figsize = (12, 6))

sns.barplot(
    data = avg_speed_by_distance_gender, 
    x = 'Event distance/length', 
    y = 'Athlete average speed', 
    hue = 'Athlete gender', 
    palette = 'coolwarm'
)

plt.xlabel('Event Distance/Length')
plt.ylabel('Average Speed')
plt.title('Average Speed by Event Distance/Length and Gender')
plt.legend(title = 'Gender')
plt.show()
```

```python
# Season average speed comparisons
aus_races.groupby('Season')['Athlete average speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
avg_speed_by_season = aus_races.groupby('Season')['Athlete average speed'].agg(['mean', 'count']).reset_index()

# Plot the bar plot
plt.figure(figsize = (12, 6))

sns.barplot(data = avg_speed_by_season, x = 'Season', y = 'mean', palette = 'Set2')
plt.xlabel('Season')
plt.ylabel('Average Speed')
plt.title('Average Speed by Season')
plt.show()
```
